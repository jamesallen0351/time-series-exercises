# Acquire Exercises

```python
# imports
import pandas as pd
import requests
import os
```

```python
# creating a function to get the store data
def store_data(original, name):
    if os.path.isfile(name +'.csv'):
        return pd.read_csv(name + '.csv')
    else:
        
        tlist = []
        df = pd.DataFrame()
        response = requests.get(original)
        data = response.json()
        n = data['payload']['max_page']

        for i in range(1,n+1):
            url = original +'?page='+str(i)
            response = requests.get(url)
            data = response.json()
            page_items = data['payload'][name]
            tlist += page_items
            url = original
        df = pd.DataFrame(tlist)
        df.to_csv(name + '.csv',index=False)
        return df
```

```python
# getting the 'items' from the store data using 'store_data' function
items = store_data('https://python.zgulde.net/api/v1/items', 'items')
```

```python
items.head() #check_yo_head
```

```python
# looking at the shape of the items data
items.shape
```

```python
# getting the 'stores' data using the 'store_data' function
stores = store_data('https://python.zgulde.net/api/v1/stores', 'stores')
```

```python
stores.head() # check_yo_head
```

```python
# taking a look at the shape of the 'stores' data
stores.shape
```

```python
# getting the 'sales' data using the 'store_data' function
sales  = store_data('https://python.zgulde.net/api/v1/sales', 'sales')
```

```python
sales.head() # check_yo_head
```

```python
# taking a look at the shape of the 'sales' data
sales.shape
```

```python
# merging the 'items' data with the 'sales' data
item_sales = items.merge(sales, how='left', left_on= items.item_id, right_on=sales.item )
```

```python
item_sales.head()
```

```python
item_sales.shape
```

```python
store_data = item_sales.merge(stores, how='left', left_on='store', right_on=stores.store_id)
```

```python
store_data.head()
```

```python
store_data.shape
```

```python
store_data.to_csv('complete_store_data.csv', index=False)
```
